```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import pandas as pd

# Create an empty DataFrame to store restaurant details
restaurant_df = pd.DataFrame(columns=['Restaurant Name', 'Order Items', 'Prices', 'Rating', 'Votes'])

# Create a new Edge WebDriver
driver = webdriver.Edge()

# Zomato India URL
url = "https://www.zomato.com/india"
driver.get(url)

# Wait for the page to load
wait = WebDriverWait(driver, 10)
try:
    wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="sc-bke1zw-1 fTtEw"]/a')))
except TimeoutException:
    print("Page did not load within the specified time.")

# Find all the restaurant links for different cities using the appropriate XPath
cities = driver.find_elements(By.XPATH, '//div[@class="sc-bke1zw-1 fTtEw"]/a')
cities_1 = driver.find_elements(By.XPATH, '//div[@class="sc-bke1zw-1 bcVYKA"]/a')

# Create a list to store the href values
hrefs = []

# Iterate through the found links and extract href attributes
for city, city_1 in zip(cities, cities_1):
    link_0 = city.get_attribute('href')
    link_1 = city_1.get_attribute('href')
    hrefs.extend([link_0, link_1])

# Open each link in the browser
for link in hrefs:
    driver.get(link)
    print(link)
    
    try:
        # Check if "Dining Out" element is present
        dining_out_element = wait.until(
            EC.presence_of_element_located((By.XPATH, '//div[@class="sc-gxbSSY fsqyAa" and text()="Dining Out"]'))
        )
        
        # If present, click it and do something on the page, e.g., scrape data
        dining_out_element.click()

        for j in range(10, 13):  # Update the range according to your needs
            for i in range(1, 4):  # Update the range according to your needs
                xpath = '//*[@id="root"]/div/div[' + str(j) + ']/div/div[' + str(i) + ']/div/div/a[2]'
                
                # Use find_elements to find multiple restaurant elements
                restaurant_elements = wait.until(
                    EC.presence_of_all_elements_located((By.XPATH, xpath))
                )
                
                # Iterate through each restaurant element
                for restaurant_element in restaurant_elements:
                    rest_face = restaurant_element.get_attribute('href')
                    driver.execute_script("window.open('" + rest_face + "', '_blank');")
                    driver.switch_to.window(driver.window_handles[1])
                    
                    # Example: Add the scraped data to the DataFrame
                    try:
                        rest_name = wait.until(
                            EC.presence_of_element_located((By.CLASS_NAME, "sc-7kepeu-0.sc-iSDuPN.fwzNdh"))
                        )
                        print(rest_name.text)
                        
                        # Perform scraping or other actions here
                        restaurant_df = pd.concat([restaurant_df, pd.DataFrame({
                            'Restaurant Name': [rest_name.text],
                            'Order Items': ['-'],
                            'Prices': ['-'],
                            'Rating': ['-'],
                            'Votes': ['-'],
                        })], ignore_index=True)
                    except NoSuchElementException:
                        print("Restaurant name not found on this page. Skipping...")
                    
                    # Close the tab after scraping data
                    driver.close()
                    
                    # Switch back to the main city page
                    driver.switch_to.window(driver.window_handles[0])
                    
            # Move back to the main city page after iterating through the restaurants
            driver.back()

    except NoSuchElementException:
        print('Dining Out element not found on this page. Skipping...', link)

# Save the DataFrame to a CSV file
restaurant_df.to_csv('restaurant_details.csv', index=False)

# Close the browser
driver.quit()
```
